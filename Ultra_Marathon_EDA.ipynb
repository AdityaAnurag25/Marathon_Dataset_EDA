Filtering 50km and 50mi races which took place in USA in the year 2022.
First we filter by distance and year.
Then we filter by Place of event.

```python
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2022)]
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

Now we combine all the filters in order to get the desired dataset.

```python
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2022) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

After combining all the filters, now we create a new dataframe which will contain the desired dataset.

```python
df2 = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2022) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head(3)
```

Now we remove the country name from the event name since this dataframe contains only the events which took place in USA in the year 2022.

```python
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

Since the age of the atheletes is not given in the dataset, we calculate the age of the athletes using their year of birth and the year in which the event took place, i.e, 2022.

```python
df2['athlete_age'] = 2022 - df2['Athlete year of birth']
```

After calculating athlete age, we now clean the athlete performance column by removing the 'h' at the end of the entries. This is being done considering that marathon timings are generally noted in hours and it is not necessary to mention 'h' at the end of the entries

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

Now we remove certain columns which are not useful for this analysis, namely, 'Athlete club', 'athlete country', 'athlete year of birth' and 'athlete age category'

```python
df2 = df2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'], axis = 1)
```

Every dataset has rows which do not contain any values, i.e, NULL. Now we remove those NULL values from the dataset.

```python
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna() == 1]
```

```python
df2 = df2.dropna()
```

```python
df2.dtypes
```

We change the datatype of certain columns which will enable us to do calculations on these columns later when the analysis is done.

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

We rename the columns into a more coding friendly syntax for the ease of our work.

```python
df2 = df2.rename(columns = {'Year of event': 'year',
                            'Event dates': 'race_date',
                            'Event name': 'race_name',
                            'Event distance/length': 'race_length',
                            'Event number of finishers': 'race_number_of_finishers',
                            'Athlete performance': 'athlete_performance',
                            'Athlete gender': 'athlete_gender',
                            'Athlete average speed': 'athlete_average_speed',
                            'Athlete ID': 'athlete_id',
                            })
```

Questions to be answered from the dataset
1. Difference in speed for the 50k, 50mi male to female.
2. What age groups are the best in the 50m Race (20+ races min)
3. What age groups are the best in the 50m Race (20+ races min)
4. Verify whether participants are slower in the summer as compared to winter.

```python
#Difference in speed for the 50k, 50mi male to female
df2.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
#What age groups are the best in the 50m Race (20+ races min)
df2.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False).query('count>19').head(10)
```

```python
#What age groups are the worst in the 50m Race (10+ races min)
df2.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = True).query('count>9').head(10)
```

```python
# Verify whether participants are slower in the summer as compared to winter.
# spring 3-5
# summer 6-8
# fall 9-11
# winter 12-2
# split between two decimals
```

```python
df2['race_month'] = df2['race_date'].str.split('.').str.get(1).astype(int)
```

```python
df2['race_season'] = df2['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df2.head(2)
```

```python
df2.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```

```python
df2.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```
